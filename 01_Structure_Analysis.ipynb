```python
from mp_api.client import MPResters
import py3Dmol

# 1. 請填入你的 Materials Project API Key (去官網 Dashboard 複製)
API_KEY = "YOUR_API_KEY_HERE"

# 2. 從資料庫抓取鋁 (Al) 的結構，ID 為 mp-134
with MPRester(API_KEY) as mpr:
    structure = mpr.get_structure_by_material_id("mp-134")

print(f"成功抓取材料：{structure.composition.reduced_formula}")
print(f"晶系: {structure.get_space_group_info()}")

# 3. 3D 視覺化
cif_str = structure.to(fmt="cif")
viewer = py3Dmol.view(width=400, height=400)
viewer.addModel(cif_str, "cif")
viewer.setStyle({'sphere': {'colorscheme': 'Jmol', 'scale': 0.3}, 'stick': {'radius': 0.1}})
viewer.addUnitCell()
viewer.zoomTo()
viewer.show()
```

```python
from mp_api.client import MPRester
import py3Dmol


API_KEY = "og8XLgtAMAhivvpdnoaifDGlRikb03bL"

with MPRester(API_KEY) as mpr:
    data = mpr.materials.summary.search(
        material_ids=["mp-804"], 
        fields=["structure", "band_gap", "formula_pretty"]
    )[0]

structure = data.structure
gap = data.band_gap

print(f"材料: {data.formula_pretty}")
print(f"能隙 (Band Gap): {gap:.2f} eV")


# 判斷導電性
if gap == 0:
    print("這是金屬 (Metal)")
elif 0 < gap <= 3.0:
    print(f"這是典型的半導體")
elif 3.0 < gap <= 5.0:
    print(f"這是寬能隙半導體")
else:
    print(f"這是絕緣體")

# 3D 視覺化 (這次我們換個顏色風格)
cif_str = structure.to(fmt="cif")
viewer = py3Dmol.view(width=400, height=400)
viewer.addModel(cif_str, "cif")
viewer.setStyle({'sphere': {'colorscheme': 'carbon', 'scale': 0.3}, 'stick': {'radius': 0.1}})
viewer.addUnitCell()
viewer.zoomTo()
viewer.show()
```
